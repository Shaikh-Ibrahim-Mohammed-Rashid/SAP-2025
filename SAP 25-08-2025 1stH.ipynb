```python
# Simple neural network with 1 input, 1 weight, 1 bias
# Predict y = 2 * x

# Initialize weight and bias
weight = 0.5     # starting guess
bias = 0.1       # starting guess

# Input and target output
x = 2.0          # input value
y_true = 4.0     # true output value (2 * 2)

# Learning rate
lr = 0.1

# Training loop
for epoch in range(10):  # Run for 10 steps
    # ----------------------------
    # 1. Forward Pass
    y_pred = (x * weight) + bias

    # 2. Calculate Loss (Mean Squared Error)
    loss = (y_pred - y_true) ** 2

    # 3. Calculate Gradients
    dL_dy_pred = 2 * (y_pred - y_true)
    dL_dweight = dL_dy_pred * x
    dL_dbias = dL_dy_pred * 1

    # 4. Update Weight and Bias
    weight = weight - lr * dL_dweight
    bias = bias - lr * dL_dbias

    # Print progress
    print(f"Epoch {epoch+1}")
    print(f"  Prediction: {y_pred:.4f}")
    print(f"  Loss: {loss:.4f}")
    print(f"  Updated Weight: {weight:.4f}")
    print(f"  Updated Bias: {bias:.4f}")
    print("-" * 30)
```
